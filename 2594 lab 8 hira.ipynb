```python
!pip install numpy
!pip install pandas
import numpy as np
import pandas as pd
from  scipy import stats
#representing data as datafrmae
df = pd.read_csv('employee_details.csv')
df1=pd.read_csv('employee_salaries.csv')
print(df)
print(df1)
df2=pd.merge(df,df1,on="Employee_ID",how="outer")
print("the merging result")
print(df2)
#now remove the missing values from the columns named names and missing values
df3=df2.dropna(subset=['Name','Salary'])
print(df3)
```

```python
!pip install numpy
!pip install pandas
import numpy as np
import pandas as pd
from  scipy import stats
#representing data as datafrmae
df = pd.read_csv('employee_details.csv')
df1=pd.read_csv('employee_salaries.csv')
print(df)
print(df1)
df2=pd.merge(df,df1,on="Employee_ID",how="outer")
print("the merging result")
print(df2)
df2['Department'].fillna('Unknown',inplace=True)
print(df2)
df2['Bonus'].fillna(df2['Bonus'].mean(),inplace=True)
print(df2)
df2['Salary'].fillna(df2['Salary'].mean(),inplace=True)
print(df2)
```

```python
!pip install numpy
!pip install pandas
import numpy as np
import pandas as pd
from  scipy import stats
#representing data as datafrmae
df = pd.read_csv('employee_details.csv')
df1=pd.read_csv('employee_salaries.csv')
print(df)
print(df1)
df2=pd.merge(df,df1,on="Employee_ID",how="outer")
print("the merging result")
print(df2)
Q1=df2['Salary'].quantile(0.25);
Q3=df2['Salary'].quantile(0.75);
IQR=Q3-Q1;
df2_outliers=df2[(df2['Salary']<Q1-1.5*IQR)|(df2['Salary']>Q3+1.5*IQR)]
print(df2_outliers)
```
